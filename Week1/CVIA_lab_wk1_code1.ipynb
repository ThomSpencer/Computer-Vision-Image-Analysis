```python
# Load an image
myImg = mpimg.imread('example_image.jpg')
```

```python
# Show image
plt.figure(1)                     # Creates a figure window
plt.imshow(myImg)                 # Display the image
plt.title('Input Image')          # Add a title
plt.axis('off')                   # Hide axes
```

```python
# Resize image to half the size
# We compute the new size and use slicing
rows, cols = myImg.shape[0], myImg.shape[1]
myHalfImg = myImg[::4, ::4]        # Take every 4 pixel

plt.figure(2)
plt.imshow(myHalfImg)
plt.title('Resized Image')
plt.axis('off')
```

```python
# Check the size of the images
myImgSize = myImg.shape
myHalfImgSize = myHalfImg.shape

print("Original image size:", myImgSize)
print("Resized image size:", myHalfImgSize)
```
