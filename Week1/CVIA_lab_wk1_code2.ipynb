```python
# CVIA Lab Week1 - Code example 2
# A simple image colour conversion program.
# The code below shows how to split an image into Red, Green, and Blue
# channels using NumPy slicing.
# The code also demonstrates displaying results using subplots.

# Author: Min Wang
# Created: 1 Feb 2026

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Load image
myImg = mpimg.imread('example_image.jpg')
```

```python
# Resize image to half size (simple subsampling)
myImgResized = myImg[::2, ::2]
```

```python
# Separate colour channels
myImgRed   = myImgResized[:, :, 0]
myImgGreen = myImgResized[:, :, 1]
myImgBlue  = myImgResized[:, :, 2]
```

```python
# Compute grayscale version of the original image
# Using standard luminance formula
myImgGray = (
    0.299 * myImg[:, :, 0] +
    0.587 * myImg[:, :, 1] +
    0.114 * myImg[:, :, 2]
)
```

```python
# Subplot example
plt.figure(7, figsize=(10, 6))

plt.subplot(2, 3, 1)
plt.imshow(myImg)
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(myImgResized)
plt.title('Resized Image')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(myImgRed, cmap='gray')
plt.title('Red Channel')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(myImgGray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(myImgGreen, cmap='gray')
plt.title('Green Channel')
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(myImgBlue, cmap='gray')
plt.title('Blue Channel')
plt.axis('off')

plt.tight_layout()
plt.show()
```
