```python
# CVIA Lab Week 3 - Code Example 2: Edge detection (Sobel, Canny)
# Author: Min Wang
# Created: 2026/02/20

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# --------------------------------------------------
# Working with the barcode image
# --------------------------------------------------

# Load barcode image
myImgBarcode = cv2.imread('barcode_cropped.jpg')
myImgBarcode = cv2.cvtColor(myImgBarcode, cv2.COLOR_BGR2RGB)

# Convert to grayscale
myImgGrayBarcode = cv2.cvtColor(myImgBarcode, cv2.COLOR_RGB2GRAY)

# --------------------------------------------------
# Sobel Edge Detection
# --------------------------------------------------
sobelx = cv2.Sobel(myImgGrayBarcode, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(myImgGrayBarcode, cv2.CV_64F, 0, 1, ksize=3)

myImgGrayBarcodeSobel = cv2.magnitude(sobelx, sobely)
myImgGrayBarcodeSobel = cv2.convertScaleAbs(myImgGrayBarcodeSobel)

# --------------------------------------------------
# Canny Edge Detection
# --------------------------------------------------
myImgGrayBarcodeCanny = cv2.Canny(myImgGrayBarcode, 100, 200)


# Plot the figures
plt.figure(figsize=(10, 6))

plt.subplot(1, 4, 1)
plt.imshow(myImgBarcode)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(myImgGrayBarcode, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(myImgGrayBarcodeSobel, cmap='gray')
plt.title('Sobel')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(myImgGrayBarcodeCanny, cmap='gray')
plt.title('Canny')
plt.axis('off')

```
