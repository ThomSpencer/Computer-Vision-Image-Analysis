```python
# CVIA Lab Week 3 - Code Example 1: Histogram equalisation
# Author: Min Wang
# Created: 2026/02/20

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# --------------------------------------------------
# Load image (OpenCV loads as BGR by default)
# --------------------------------------------------
myImg = cv2.imread('barcode_cropped.jpg')
myImg = cv2.cvtColor(myImg, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(14, 10))

# Original Image
plt.subplot(1, 5, 1)
plt.imshow(myImg)
plt.title('Original Image')
plt.axis('off')

# Separate RGB channels
myImgRed   = myImg[:, :, 0]
myImgGreen = myImg[:, :, 1]
myImgBlue  = myImg[:, :, 2]

plt.subplot(1, 5, 2)
plt.imshow(myImgRed, cmap='gray')
plt.title('Red Channel')
plt.axis('off')

plt.subplot(1, 5, 3)
plt.imshow(myImgGreen, cmap='gray')
plt.title('Green Channel')
plt.axis('off')

plt.subplot(1, 5, 4)
plt.imshow(myImgBlue, cmap='gray')
plt.title('Blue Channel')
plt.axis('off')

# Grayscale image
myImgGray = cv2.cvtColor(myImg, cv2.COLOR_RGB2GRAY)

plt.subplot(1, 5, 5)
plt.imshow(myImgGray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
```

```python
# --------------------------------------------------
# Histograms
# --------------------------------------------------
plt.figure(figsize=(14, 3))

plt.subplot(1, 4, 1)
plt.hist(myImgRed.ravel(), bins=256, range=(0,256))
plt.title('Red Histogram')

plt.subplot(1, 4, 2)
plt.hist(myImgGreen.ravel(), bins=256, range=(0,256))
plt.title('Green Histogram')

plt.subplot(1, 4, 3)
plt.hist(myImgBlue.ravel(), bins=256, range=(0,256))
plt.title('Blue Histogram')

plt.subplot(1, 4, 4)
plt.hist(myImgGray.ravel(), bins=256, range=(0,256))
plt.title('Gray Histogram')
```

```python
# --------------------------------------------------
# Histogram Equalisation
# --------------------------------------------------
myImgRedHistEq   = cv2.equalizeHist(myImgRed)
myImgGreenHistEq = cv2.equalizeHist(myImgGreen)
myImgBlueHistEq  = cv2.equalizeHist(myImgBlue)
myImgGrayHistEq  = cv2.equalizeHist(myImgGray)
```

```python
# --------------------------------------------------
# Display equalised channels
# --------------------------------------------------
plt.figure(figsize=(14, 3))

plt.subplot(1, 4, 1)
plt.imshow(myImgRedHistEq, cmap='gray')
plt.title('Red (Hist Eq)')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(myImgGreenHistEq, cmap='gray')
plt.title('Green (Hist Eq)')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(myImgBlueHistEq, cmap='gray')
plt.title('Blue (Hist Eq)')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(myImgGrayHistEq, cmap='gray')
plt.title('Grayscale (Hist Eq)')
plt.axis('off')
```

```python
# --------------------------------------------------
# Histograms after equalisation
# --------------------------------------------------
plt.figure(figsize=(14, 3))

plt.subplot(1, 4, 1)
plt.hist(myImgRedHistEq.ravel(), bins=256, range=(0,256))
plt.title('Image Histogram')

plt.subplot(1, 4, 2)
plt.hist(myImgGreenHistEq.ravel(), bins=256, range=(0,256))
plt.title('Image Histogram')

plt.subplot(1, 4, 3)
plt.hist(myImgBlueHistEq.ravel(), bins=256, range=(0,256))
plt.title('Image Histogram')

plt.subplot(1, 4, 4)
plt.hist(myImgGrayHistEq.ravel(), bins=256, range=(0,256))
plt.title('Image Histogram')
```

```python
# --------------------------------------------------
# Combine equalised RGB channels
# --------------------------------------------------
myImgHistEq = np.zeros_like(myImg)
myImgHistEq[:, :, 0] = myImgRedHistEq
myImgHistEq[:, :, 1] = myImgGreenHistEq
myImgHistEq[:, :, 2] = myImgBlueHistEq

plt.figure(figsize=(5, 3))
plt.imshow(myImgHistEq)
plt.title('After Hist Eq')
plt.axis('off')

plt.tight_layout()
plt.show()
```
